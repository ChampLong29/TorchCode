```python
def cross_entropy_loss(logits, targets):
    max_vals = logits.max(dim=-1, keepdim=True).values
    shifted = logits - max_vals
    max_vals = max_vals.squeeze(-1)
    log_sum_exp = torch.log(torch.sum(torch.exp(shifted), dim=-1))
    targets = targets.unsqueeze(-1)
    target_logits = logits.gather(dim=-1, index=targets)
    loss = -target_logits + max_vals + log_sum_exp
    return loss.mean()
```

```python
# Demo
logits = torch.randn(4, 10)
targets = torch.randint(0, 10, (4,))
print('Loss:', cross_entropy_loss(logits, targets).item())
print('Ref: ', torch.nn.functional.cross_entropy(logits, targets).item())
```
