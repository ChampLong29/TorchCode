### Interview Context
"Sparse vs dense rewards — how do you design process rewards?"
Outcome rewards are sparse (only at end). Process rewards give intermediate
feedback (e.g., format correctness, tool-call success, partial credit).
DeepSeek-R1 uses outcome-only; OpenAI o1-style uses process reward models (PRM).
