### Rules
- Find all `<think>`/`</think>` pairs and mark tokens between them
- Find all `<action>`/`</action>` pairs and mark tokens between them
- Support multiple think/action sections
- Tokens that are the special tags themselves are NOT included in either mask

### Interview Context
"How do you compute loss on reasoning vs action tokens separately?"
→ Reasoning tokens can use a different loss weight, or be excluded from the RL update
(KL penalty only on action tokens, not reasoning).
