
### Rules
- Call `env.reset()` to get initial state
- Loop: compute action from policy, step env, store transition dict
- Stop when `done=True` or `max_steps` reached
- Each transition dict must contain all 5 keys: state, action, reward, next_state, done

### Example
```python
>>> traj = rollout(my_policy, my_env, max_steps=10)
>>> len(traj)
4  # terminated after 4 steps
>>> traj[0]
{"state": 0.0, "action": 1.0, "reward": 1.0, "next_state": 1.0, "done": False}
```
