```python
        response_text: str, model output possibly containing a tool call
        tools: dict of {name: callable(**kwargs) -> result}
    Returns:
        (tool_name, result) tuple, or None if no tool call detected
    """
    ...
```

### Rules
- Support JSON-format tool calls: `{"tool": "name", "args": {...}}`
- Parse `response_text`, match tool name against `tools` dict
- Call the matched tool with parsed arguments, return `(name, result)`
- Return `None` if no valid tool call found

### Example
```python
>>> tools = {"add": lambda a, b: a + b}
>>> execute_tool_call('{"tool": "add", "args": {"a": 3, "b": 5}}', tools)
('add', 8)
>>> execute_tool_call('Just a normal reply.', tools)
None
```
