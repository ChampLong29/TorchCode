```python
    ...
```

### Rules
- Policy head: `Linear(d_model, hidden) -> ReLU -> Linear(hidden, action_dim)`
- Value head: `Linear(d_model, hidden) -> ReLU -> Linear(hidden, 1)`
- Both heads should be `nn.Module` instances (use `nn.Sequential`)
- Initialize weights with Kaiming uniform (He init)

### Interview Context
Interviewers ask: "Why separate heads? How to freeze base while training heads for RL?"
Answer: Freezing base preserves pretrained capabilities while heads learn task-specific
policy. This enables KL-constrained RL: base (reference) vs policy divergence.
