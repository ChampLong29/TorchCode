```python
import random

def rand16():
    return random.randint(0, 2 ** 16 - 1)

def uniform_rand(n: int) -> int:
    while True:
        target_bits = n.bit_length()
        base_bits = 16
        total_bits = 0
        x = 0
        while total_bits < target_bits:
            x = (x << base_bits) | rand16()
            total_bits += base_bits
        if x < n:
            return x

        

```

```python
def sample(N: int, x: int) :
    ans = set()
    while len(ans) < x:
        ans.add(uniform_rand(N))

    return list(ans)
```
