### Rules
- Each epoch: collect N rollouts, compute discounted returns, PPO loss, backprop + step
- The policy should output a scalar action given a state: `policy(state) -> action`
- Compute advantages as (returns - mean(returns)) / (std(returns) + 1e-8)
- Use a simplified PPO loss: `min(r * A, clip(r, 1-eps, 1+eps) * A)` where r is ratio (you can approximate with new/old output)
- Return list of mean losses tracking training progress

````python
#  Test your implementation

class SimplePolicy(nn.Module):
    def __init__(self):
        super().__init__()
        self.scale = nn.Parameter(torch.tensor([0.5]))
    def forward(self, x):
        return self.scale.abs() * x

class SimpleEnv:
    def __init__(self): self.i = 0
    def reset(self): self.i = 0; return torch.tensor(1.0)
    def step(self, action):
        self.i += 1
        s = torch.tensor(float(self.i + 1))
        return s, -float(action), self.i >= 3

policy = SimplePolicy()
ref_policy = SimplePolicy()
ref_policy.load_state_dict(policy.state_dict())
opt = torch.optim.SGD(policy.parameters(), lr=0.01)

losses = train_rl_pipeline(policy, ref_policy, SimpleEnv(), opt,
                           epochs=3, rollouts_per_epoch=1)
print("Losses:", losses)
assert len(losses) == 3
print(" Passed!")
````
